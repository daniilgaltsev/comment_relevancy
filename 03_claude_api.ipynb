```python
inp = 2399
out = 280
n = 5000
def price(inp, out, n, p=2):
    price_batch = (inp * 0.25 + out * 1.25) / 2 / 1000000
    price_cache = (inp * 0.25 * 0.9 + out * 1.25) / 1000000
    price_batch_cache = (inp * 0.25 * 0.9 + out * 1.25) / 2 / 1000000
    return round(price_cache * n, p), round(price_batch * n, p), round(price_batch_cache * n, p)
```

```python
print("Large", price(2399, 150, 5000), price(2399, 280, 5000), price(2399, 400, 5000))
print("Small", price(978, 150, 5000), price(978, 280, 5000), price(978, 400, 5000))
```
